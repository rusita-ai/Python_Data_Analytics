# V. 할당 연산자(Assignment Operators)

```
koo = 88

print(koo)
```

> ## 1) +=

```
# koo = koo + 11

koo += 11
print(koo)
```

> ## 2) -=

```
# koo = koo - 11

koo -= 11
print(koo)
```

> ## 3) *=

```
# koo = koo * 10

koo *= 10
print(koo)
```

> ## 4) /=

```
# koo = koo / 10

koo /= 10
print(koo)
```

> ## 5) //=

```
# koo = koo // 10     

koo //= 10 
print(koo)
```

> ## 6) %=

```
# koo = koo % 3

koo %= 3
print(koo)
```

> ## 7) **=

```
# koo = koo ** 2

koo **= 2
print(koo)
```
